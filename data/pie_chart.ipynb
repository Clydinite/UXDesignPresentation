```python
import matplotlib.pyplot as plt

# 1. 強制重設為原始預設樣式
plt.style.use('default')

labels = ['1-3', '4-6', '7-10']
sizes = [1, 5, 1]

# 2. 建立畫布時確保背景為白色
plt.figure(figsize=(8, 6), facecolor='white')
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# 加入標題與圖例 ($N=7$ 使用 LaTeX 格式)
plt.title('Motivation to Leave Home ($N = 7$)')
plt.legend(loc="upper right")

plt.tight_layout()

# 3. 存檔時強制指定白底且關閉透明度
plt.savefig('survey_results_plot.png', dpi=300, facecolor='white', transparent=False)
plt.show()
```
